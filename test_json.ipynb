```python
import re
import json

def test(response):
    try:
        response = response.strip()
        match = re.search(r"\{.*\}", response, re.DOTALL)
        if not match:
            return {'has_quest': False, 'quest_description': '', 'item_name': ''}
        json_str = match.group(0)
        print(json_str)
        # Ensure keys are quoted - only match keys after { or ,
        json_str = re.sub(r'([{,]\s*)(\w+)(?=\s*:)', r'\1"\2"', json_str)
        print(json_str)
        # Replace single True/False with proper JSON booleans
        json_str = re.sub(r':\s*True', ': true', json_str)
        print(json_str)
        json_str = re.sub(r':\s*False', ': false', json_str)
        print(json_str)
        # Fill empty values
        json_str = re.sub(r':\s*([,}])', r': ""\1', json_str)
        print(json_str)
        result = json.loads(json_str)
        return {
            'has_quest': bool(result.get('has_quest', False)),
            'quest_description': result.get('quest_description', ''),
            'item_name': result.get('item_name', '')
        }
    except Exception as e:
        print(f"Failed to parse quest analysis: {e}, response: {response}\n")
        return {'has_quest': False, 'quest_description': '', 'item_name': ''}

responses = [
    """ {has_quest: false, quest_description: , item_name: } """,
    """ {has_quest: true, quest_description: Retrouver l'âme perdue du Clock d'Âme Éternel de Veyra pour revitaliser les Clocks et soulager son âme gélée., item_name: } """
]

for response in responses:
    result = test(response)
    print(f"Result: {result}")
```
